```python
import pandas as pd
import matplotlib.pyplot as plt
from textblob import TextBlob

# Assuming 'Text' is the column containing text data in your DataFrame
# Replace 'Text' with the actual column name in your dataset

# Example DataFrame
data = {
    'Text': ['I love this product!', 'It was okay.', 'Not satisfied with the service.', 'Amazing experience.'],
    'Year': [2020, 2020, 2021, 2021]
}

df = pd.DataFrame(data)

# Function to get sentiment polarity
def get_sentiment(text):
    analysis = TextBlob(text)
    # Assign sentiment polarity as 'positive', 'neutral', or 'negative'
    if analysis.sentiment.polarity > 0:
        return 'positive'
    elif analysis.sentiment.polarity == 0:
        return 'neutral'
    else:
        return 'negative'

# Apply sentiment analysis and create a new 'Sentiment' column
df['Sentiment'] = df['Text'].apply(get_sentiment)

# Calculate the percentage of each sentiment
sentiment_percentage = df['Sentiment'].value_counts(normalize=True) * 100

# Create a pie chart for sentiment distribution
plt.figure(figsize=(8, 8))
plt.pie(sentiment_percentage, labels=sentiment_percentage.index, autopct='%1.1f%%', startangle=90, colors=['lightgreen', 'lightblue', 'lightcoral'])
plt.title('Sentiment Distribution')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from textblob import TextBlob

# Assuming 'Text' is the column containing text data in your DataFrame
# Replace 'Text' with the actual column name in your dataset

# Example DataFrame with 'Medals' column
data = {
    'Text': ['I love this product!', 'It was okay.', 'Not satisfied with the service.', 'Amazing experience.'],
    'Medals': [2, 1, 0, 3]
}

df = pd.DataFrame(data)

# Function to get sentiment based on medal count
def get_sentiment_with_medals(medals, text):
    if medals == 1:
        return 'neutral'
    elif medals == 0:
        return 'not negative'
    elif medals > 2:
        return 'positive'
    else:
        # Default to neutral if conditions are not met
        return 'neutral'

# Apply sentiment analysis based on medal count and create a new 'Sentiment' column
df['Sentiment'] = df.apply(lambda row: get_sentiment_with_medals(row['Medals'], row['Text']), axis=1)

# Calculate the percentage of each sentiment
sentiment_percentage = df['Sentiment'].value_counts(normalize=True) * 100

# Define custom colors
colors = {'not negative': 'orange', 'positive': 'green', 'neutral': 'blue', 'negative': 'red'}

# Create a pie chart with custom colors
plt.figure(figsize=(8, 8))
explode = (0.1, 0, 0)  # Extract positive piece by 10%
plt.pie(sentiment_percentage, labels=sentiment_percentage.index, autopct='%1.1f%%', startangle=90, colors=[colors.get(s, s) for s in sentiment_percentage.index], explode=explode)
plt.title('Sentiment Distribution')
plt.show()
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Example DataFrame with relevant columns (replace this with your actual dataset)
data = {
    'Sport': ['Judo', 'Judo', 'Judo', 'Judo', 'Judo', 'Taekwondo', 'Taekwondo', 'Taekwondo', 'Taekwondo', 'Taekwondo'],
    'NOC': ['JPN', 'FRA', 'RUS', 'JPN', 'KOR', 'KOR', 'CHN', 'USA', 'CHN', 'KOR'],
    'Gender': ['M', 'M', 'F', 'M', 'F', 'M', 'F', 'M', 'F', 'M'],
    'Height': [170, 180, 160, 175, 165, 175, 160, 180, 165, 170],
    'Year': [2000, 2004, 2008, 2012, 2016, 2000, 2004, 2008, 2012, 2016]
}

olympics_df = pd.DataFrame(data)

# Function to get the top 3 participating nations for a given sport
def top_participating_nations(df, sport, top_n=3):
    sport_df = df[df['Sport'] == sport]
    top_nations = sport_df['NOC'].value_counts().head(top_n).index.tolist()
    return sport_df[sport_df['NOC'].isin(top_nations)]

# Find the top 3 participating nations for Judo and Taekwondo
top_judo_df = top_participating_nations(olympics_df, 'Judo')
top_taekwondo_df = top_participating_nations(olympics_df, 'Taekwondo')

# Combine the datasets for analysis
combined_df = pd.concat([top_judo_df, top_taekwondo_df])

# Plot the distribution of height based on gender for the top 3 countries in Judo and Taekwondo
plt.figure(figsize=(10, 6))
sns.boxplot(x='NOC', y='Height', hue='Gender', data=combined_df)
plt.title('Height Distribution Based on Gender for Top 3 Countries in Judo and Taekwondo (1960-2016)')
plt.xlabel('Country')
plt.ylabel('Height (cm)')
plt.show()
```
